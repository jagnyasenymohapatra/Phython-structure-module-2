# Stastics Part2 Theory

### 1. What is hypothesis testing in statistics?  
Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample of data to infer that a certain condition is true for the entire population. It involves formulating two opposing hypotheses, conducting a test, and analyzing the results to make a decision.



### 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?  
The null hypothesis (\(H_0\)) is a statement that assumes there is no significant effect or difference in a population. It represents the default assumption that nothing has changed.  

The alternative hypothesis (\(H_1\) or \(H_a\)) is the opposite of the null hypothesis, stating that there is a significant effect or difference.  

The key difference is that the null hypothesis assumes no change, while the alternative hypothesis suggests that there is a meaningful change.


### 3. What is the significance level in hypothesis testing, and why is it important?  
The significance level (\(\alpha\)) is the probability of rejecting the null hypothesis when it is actually true. It is typically set at 0.05 (5%) or 0.01 (1%). It is important because it defines the threshold for statistical significance, helping to control the likelihood of making a Type I error (false positive).



### 4. What does a P-value represent in hypothesis testing?  
The P-value is the probability of obtaining a test statistic at least as extreme as the one observed, assuming the null hypothesis is true. It quantifies the strength of evidence against the null hypothesis.



### 5. How do you interpret the P-value in hypothesis testing?  
- If the P-value is less than the significance level (\(\alpha\)), we reject the null hypothesis, suggesting that the observed result is statistically significant.  
- If the P-value is greater than \(\alpha\), we fail to reject the null hypothesis, meaning there is not enough evidence to support the alternative hypothesis.



### 6. What are Type 1 and Type 2 errors in hypothesis testing?  
- A **Type 1 error** occurs when the null hypothesis is true but is incorrectly rejected (false positive).  
- A **Type 2 error** occurs when the null hypothesis is false but is incorrectly accepted (false negative).  

Type 1 errors are controlled by the significance level (\(\alpha\)), while Type 2 errors depend on the power of the test.



### 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?  
- A **one-tailed test** examines whether a parameter is greater than or less than a specific value in one direction.  
- A **two-tailed test** checks for differences in both directions, meaning it tests whether the parameter is either significantly higher or lower than expected.  

One-tailed tests are more powerful in detecting an effect in one direction, while two-tailed tests are more conservative.


### 8. What is the Z-test, and when is it used in hypothesis testing?  
A **Z-test** is a statistical test used to determine if there is a significant difference between sample and population means when the population variance is known and the sample size is large (\(n \geq 30\)). It is used for comparing means and proportions.


### 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?  
The **Z-score** is calculated using the formula:  
\[
Z = \frac{X - \mu}{\sigma}
\]  
where \(X\) is the observed value, \(\mu\) is the population mean, and \(\sigma\) is the population standard deviation.  

It represents how many standard deviations an observation is from the mean.


### 10. What is the T-distribution, and when should it be used instead of the normal distribution?  
The **T-distribution** is a probability distribution that is used when the population standard deviation is unknown and the sample size is small (\(n < 30\)). It is similar to the normal distribution but has heavier tails.

---

### 11. What is the difference between a Z-test and a T-test?  
A **Z-test** is used when the population variance is known and the sample size is large, whereas a **T-test** is used when the population variance is unknown and the sample size is small.



### 12. What is the T-test, and how is it used in hypothesis testing?  
A **T-test** is a statistical test used to compare the means of two groups. It helps determine whether the difference between the sample means is statistically significant.



### 13. What is the relationship between Z-test and T-test in hypothesis testing?  
Both tests compare sample and population parameters, but the Z-test is used for large samples with known variance, while the T-test is used for small samples with unknown variance.



### 14. What is a confidence interval, and how is it used to interpret statistical results?  
A **confidence interval (CI)** is a range of values within which the true population parameter is expected to lie with a certain level of confidence (e.g., 95%). It helps quantify the uncertainty in estimation.



### 15. What is the margin of error, and how does it affect the confidence interval?  
The **margin of error** is the range within which the true population parameter is expected to fall. A smaller margin of error means higher precision, while a larger margin of error increases uncertainty.


### 16. How is Bayes' Theorem used in statistics, and what is its significance?  
Bayes' Theorem calculates the probability of an event based on prior knowledge. It is significant in updating probabilities as new evidence is introduced.


### 17. What is the Chi-square distribution, and when is it used?  
The **Chi-square distribution** is used for categorical data analysis, particularly in goodness-of-fit tests and tests of independence.



### 18. What is the Chi-square goodness-of-fit test, and how is it applied?  
It is a statistical test used to determine whether observed categorical data matches expected distributions.



### 19. What is the F-distribution, and when is it used in hypothesis testing?  
The **F-distribution** is used in variance analysis, such as ANOVA and the F-test, to compare multiple population variances.


### 20. What is an ANOVA test, and what are its assumptions?  
The **Analysis of Variance (ANOVA)** test compares means across multiple groups to determine if there are significant differences. Assumptions include:  
- Independence of observations  
- Normality of the data  
- Homogeneity of variances  


### 21. What are the different types of ANOVA tests?  
- **One-way ANOVA**: Compares means across one factor.  
- **Two-way ANOVA**: Examines the effect of two factors.  
- **Repeated measures ANOVA**: Analyzes data from repeated observations on the same subjects.  



### 22. What is the F-test, and how does it relate to hypothesis testing?  
The **F-test** compares the variances of two or more groups. It is commonly used in ANOVA and regression analysis to test the significance of group differences.

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import numpy as np
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, sample_size):


    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Interpret the results
    alpha = 0.05  # Significance level
    if p_value < alpha:
        interpretation = (
            "Reject the null hypothesis. There is sufficient evidence to conclude "
            "that the sample mean is significantly different from the population mean."
        )
    else:
        interpretation = (
            "Fail to reject the null hypothesis. There is not enough evidence to "
            "conclude that the sample mean is significantly different from the population mean."
        )

    return z_score, p_value, interpretation


# Example usage
sample_mean = 75
population_mean = 70
population_std = 10
sample_size = 50

z_score, p_value, interpretation = z_test(sample_mean, population_mean, population_std, sample_size)

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.3f}")
interpretation
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu_0, mu_1, sigma, n, alpha):


    z_alpha = norm.ppf(1 - alpha / 2)  # Two-tailed test
    z_beta = (mu_0 - mu_1) / (sigma / np.sqrt(n)) + z_alpha

    beta = norm.cdf(z_beta)
    power = 1- beta
    x = np.linspace(mu_0-4*sigma/np.sqrt(n), mu_1+4*sigma/np.sqrt(n), 500)

    # Plot the distributions
    plt.figure(figsize=(10,6))

    plt.plot(x, norm.pdf(x, loc=mu_0, scale=sigma / np.sqrt(n)), label=f'H0: ={mu_0}')
    plt.plot(x, norm.pdf(x, loc=mu_1, scale=sigma / np.sqrt(n)), label=f'H1: ={mu_1}')


    # Critical regions
    x_crit = np.linspace(mu_0 + z_alpha * sigma / np.sqrt(n), mu_0 + 4 * sigma / np.sqrt(n), 100)
    plt.fill_between(x_crit, 0, norm.pdf(x_crit, loc=mu_0, scale=sigma / np.sqrt(n)), color="red", alpha=0.3, label='Type I error region')

    x_crit = np.linspace(mu_0 - 4 * sigma / np.sqrt(n), mu_0 - z_alpha * sigma / np.sqrt(n), 100)
    plt.fill_between(x_crit, 0, norm.pdf(x_crit, loc=mu_0, scale=sigma / np.sqrt(n)), color="red", alpha=0.3)


    x_beta_region = np.linspace(mu_0-z_alpha*sigma/np.sqrt(n),mu_0 +z_alpha*sigma/np.sqrt(n),100)
    plt.fill_between(x_beta_region, 0, norm.pdf(x_beta_region, loc=mu_1, scale=sigma/np.sqrt(n)), color = "blue", alpha =0.3, label = "Type II error Region")


    plt.axvline(mu_0 + z_alpha * sigma / np.sqrt(n), color = "black", linestyle="--", label = "Critical value")
    plt.axvline(mu_0 - z_alpha * sigma / np.sqrt(n), color = "black", linestyle="--")

    plt.title('Type I and Type II Error Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()

    print(f"Beta (Type II error rate): {beta:.3f}")
    print(f"Power of the test: {power:.3f}")



visualize_errors(mu_0=10, mu_1=12, sigma=2, n=50, alpha=0.05)
```

```python
# 6. Write a Python program to perform an independent T-test and interpret the results

import numpy as np
from scipy.stats import ttest_ind

def perform_independent_t_test(data1, data2, alpha=0.05):


    t_statistic, p_value = ttest_ind(data1, data2)

    interpretation = ""
    if p_value < alpha:
        interpretation = (
            "Reject the null hypothesis. There is sufficient evidence to "
            "suggest that the means of the two groups are significantly different."
        )
    else:
        interpretation = (
            "Fail to reject the null hypothesis. There is not enough evidence to "
            "suggest that the means of the two groups are significantly different."
        )

    results = {
        "t_statistic": t_statistic,
        "p_value": p_value,
        "interpretation": interpretation
    }
    return results


data_group1 = [15, 18, 20, 22, 17, 19, 21, 23, 16, 20]
data_group2 = [25, 28, 30, 27, 26, 29, 31, 24, 28, 32]


results = perform_independent_t_test(data_group1, data_group2)


print("Independent T-test Results:")
print(f"T-statistic: {results['t_statistic']:.3f}")
print(f"P-value: {results['p_value']:.3f}")
print(results["interpretation"])
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):

    try:
        data = np.array(data)
        if not np.issubdtype(data.dtype, np.number):
          return "Invalid input: Data must be numeric."
        n = len(data)
        if n < 2 :
          return "Invalid input: At least two data points are required for calculating confidence interval"
        mean = np.mean(data)
        std_err = np.std(data, ddof=1) / np.sqrt(n) # Sample standard deviation
        alpha = 1 - confidence
        t_crit = t.ppf(1 - alpha/2, df=n-1) # Using t-distribution for smaller samples
        lower_bound = mean - t_crit * std_err
        upper_bound = mean + t_crit * std_err
        return lower_bound, upper_bound
    except (TypeError,ValueError) as e:
        return f"Invalid input: {e}"

```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):

    try:
        data = np.array(data)
        if not np.issubdtype(data.dtype, np.number):
          return "Invalid input: Data must be numeric."
        n = len(data)
        if n < 2 :
          return "Invalid input: At least two data points are required for calculating margin of error"
        std_err = np.std(data, ddof=1) / np.sqrt(n)  # Sample standard error
        alpha = 1 - confidence
        t_crit = t.ppf(1 - alpha/2, df=n-1) # Using t-distribution for smaller samples
        margin = t_crit * std_err
        return margin
    except (TypeError,ValueError) as e:
        return f"Invalid input: {e}"

# Example usage
sample_data = [25, 28, 30, 27, 26, 29, 31, 24, 28, 32]
confidence_level = 0.95
error = margin_of_error(sample_data, confidence_level)
print(f"The margin of error at {confidence_level*100:.0f}% confidence is: {error:.2f}")
```

```python
# 11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

import numpy as np

def bayesian_inference(prior, likelihood, evidence):


    posterior = (likelihood * prior) / evidence
    return posterior

prior_biased = 0.5  # Prior probability of a biased coin
prior_fair = 0.5    # Prior probability of a fair coin


likelihood_biased = 0.75**5 * 0.25**2 # Binomial distribution

likelihood_fair = 0.5**5 * 0.5**2 # Binomial distribution

evidence = (likelihood_biased * prior_biased) + (likelihood_fair * prior_fair)

#Calculate posterior probabilities for biased and fair coins
posterior_biased = bayesian_inference(prior_biased, likelihood_biased, evidence)
posterior_fair = bayesian_inference(prior_fair, likelihood_fair, evidence)

print(f"Posterior probability of a biased coin: {posterior_biased:.3f}")
print(f"Posterior probability of a fair coin: {posterior_fair:.3f}")
```

```python
# 12. Perform a Chi-square test for independence between two categorical variables in Python

import pandas as pd
from scipy.stats import chi2_contingency

def chi_square_test(data, variable1, variable2):

    try:
        # Create a contingency table
        contingency_table = pd.crosstab(data[variable1], data[variable2])

        # Perform the Chi-square test
        chi2, p, dof, expected = chi2_contingency(contingency_table)

        return chi2, p, dof, expected
    except KeyError:
        return "Error: One or both variables not found in the DataFrame."
    except TypeError:
        return "Error: Input data must be a pandas DataFrame."
    except Exception as e:
        return f"An unexpected error occurred: {e}"


data = {'column1': ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B'],
        'column2': ['X', 'Y', 'X', 'Y', 'Y', 'X', 'X', 'Y']}
df = pd.DataFrame(data)


chi2, p, dof, expected = chi_square_test(df, 'column1', 'column2')

if isinstance(chi2, tuple):
    print(f"Chi2 Statistic: {chi2[0]}")
    print(f"P-value: {chi2[1]}")
    print(f"Degrees of Freedom: {chi2[2]}")
    print("Expected Frequencies:")
    print(chi2[3])
else:
    print(chi2)  # Print the error message if any
```

```python
# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

import pandas as pd
import numpy as np

def calculate_expected_frequencies(observed_data):

    try:
        if isinstance(observed_data, np.ndarray):
            observed = observed_data
        elif isinstance(observed_data, pd.DataFrame):
            observed = observed_data.values  # Convert to a NumPy array
        else:
            return "Invalid input: observed_data must be a NumPy array or a pandas DataFrame."

        row_totals = observed.sum(axis=1, keepdims=True)
        col_totals = observed.sum(axis=0, keepdims=True)
        grand_total = observed.sum()
        expected_frequencies = (row_totals @ col_totals) / grand_total

        return expected_frequencies
    except (ValueError, TypeError):
        return "Invalid input: observed_data must be a 2D array-like object with numeric values."

# Example usage with a NumPy array
observed = np.array([[10, 20, 30], [15, 25, 35]])
expected = calculate_expected_frequencies(observed)
print(expected)


# Example usage with a pandas DataFrame
data = {'A': [10, 15], 'B': [20, 25], 'C': [30, 35]}
df = pd.DataFrame(data)
expected_df = calculate_expected_frequencies(df)
expected_df

```

```python
# 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

import numpy as np
from scipy.stats import chisquare

def goodness_of_fit(observed, expected):


    chi2, p = chisquare(observed, expected)
    dof = len(observed) -1
    return {"chi2": chi2, "p-value": p, "dof": dof}

# Example usage:
observed_data = np.array([10, 20, 30, 25])
expected_data = np.array([15, 20, 25, 25])  # Example expected distribution

results = goodness_of_fit(observed_data, expected_data)
results

```

```python
#16.  Implement an F-test using Python to compare the variances of two random samples

import numpy as np
from scipy.stats import f

def f_test(sample1, sample2):

    # Calculate variances
    var1 = np.var(sample1, ddof=1)  # Use ddof=1 for sample variance
    var2 = np.var(sample2, ddof=1)

    # Ensure var1 >= var2 to avoid issues with the f.cdf function
    if var1 < var2:
        var1, var2 = var2, var1
        sample1, sample2 = sample2, sample1

    # Calculate F-statistic
    f_statistic = var1 / var2

    # Calculate degrees of freedom
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # Calculate p-value
    p_value = 2 * min(f.cdf(f_statistic, df1, df2), 1 - f.cdf(f_statistic, df1, df2)) # Two-tailed test

    return {"f_statistic": f_statistic, "p_value": p_value, "df1": df1, "df2": df2}

# Example usage
sample1 = [10, 12, 15, 11, 13]
sample2 = [12, 14, 17, 13, 15]

results = f_test(sample1, sample2)
print(results)

alpha = 0.05
if results["p_value"] < alpha:
  print("Reject null hypothesis. Variances are significantly different.")
else:
  print("Fail to reject null hypothesis. Variances are not significantly different.")
```

```python
# 17 .Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results

import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Sample data
data = {'group': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
        'value': [10, 12, 15, 18, 20, 22, 25, 27, 30]}

df = pd.DataFrame(data)

# Fit ANOVA model
model = smf.ols('value ~ C(group)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Display ANOVA table
print(anova_table)

# Interpretation
alpha = 0.05
p_value = anova_table['PR(>F)'][0]

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in means between at least two groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in means between the groups.")
```

```python
# 18. Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results
import statsmodels.formula.api as smf
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt

def perform_anova_and_plot(data, dependent_variable, independent_variable):
    try:
        # Fit the ANOVA model
        model = smf.ols(f'{dependent_variable} ~ C({independent_variable})', data=data).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)

        # Print the ANOVA table
        print("ANOVA Table:")
        print(anova_table)

        # Create a boxplot
        plt.figure(figsize=(8, 6))
        plt.boxplot([data[data[independent_variable] == g][dependent_variable] for g in data[independent_variable].unique()],
                    labels=data[independent_variable].unique())
        plt.title(f'Boxplot of {dependent_variable} by {independent_variable}')
        plt.ylabel(dependent_variable)
        plt.xlabel(independent_variable)
        plt.show()

        # Interpretation
        alpha = 0.05
        p_value = anova_table['PR(>F)'][0]

        if p_value < alpha:
            print("Reject the null hypothesis. There is a significant difference in means between at least two groups.")
        else:
            print("Fail to reject the null hypothesis. There is no significant difference in means between the groups.")

    except KeyError:
        print(f"Error: One or both variables '{dependent_variable}' and '{independent_variable}' not found in the DataFrame.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage (replace with your actual data)
data = {'group': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
        'value': [10, 12, 15, 18, 20, 22, 25, 27, 30]}
df = pd.DataFrame(data)

perform_anova_and_plot(df, 'value', 'group')
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.formula.api as smf

def check_anova_assumptions(data, dependent_variable, independent_variable):

    try:
        # Fit ANOVA model
        model = smf.ols(f'{dependent_variable} ~ C({independent_variable})', data=data).fit()
        residuals = model.resid  # Get residuals

        # **1. Normality Check - Shapiro-Wilk Test**
        shapiro_test = stats.shapiro(residuals)
        print("\nShapiro-Wilk Test for Normality:")
        print(f"Test Statistic = {shapiro_test.statistic:.4f}, p-value = {shapiro_test.pvalue:.4f}")
        if shapiro_test.pvalue > 0.05:
            print("Residuals appear to be normally distributed (Fail to reject H0). ")
        else:
            print("Residuals are not normally distributed (Reject H0). ")

        # **2. Homogeneity of Variance - Levene's Test**
        groups = [data[data[independent_variable] == g][dependent_variable] for g in data[independent_variable].unique()]
        levene_test = stats.levene(*groups)
        print("\nLevene's Test for Homogeneity of Variance:")
        print(f"Test Statistic = {levene_test.statistic:.4f}, p-value = {levene_test.pvalue:.4f}")
        if levene_test.pvalue > 0.05:
            print("Variances are equal across groups (Fail to reject H0). ")
        else:
            print("Variances are significantly different across groups (Reject H0). ")

    except KeyError:
        print(f"Error: One or both variables '{dependent_variable}' and '{independent_variable}' not found in the DataFrame.")
    except Exception as e:
        print(f"An error occurred: {e}")


data = {'group': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
        'value': [10, 12, 15, 18, 20, 22, 25, 27, 30]}
df = pd.DataFrame(data)

check_anova_assumptions(df, 'value', 'group')
```

```python
#20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns

def perform_two_way_anova(data, dependent_variable, factor1, factor2):

    try:
        # Fit Two-Way ANOVA Model
        model = smf.ols(f'{dependent_variable} ~ C({factor1}) + C({factor2}) + C({factor1}):C({factor2})', data=data).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)

        # Print ANOVA Table
        print("\nTWO-WAY ANOVA TABLE:")
        print(anova_table)

        # Interpretation
        alpha = 0.05
        for effect in anova_table.index:
            p_value = anova_table.loc[effect, 'PR(>F)']
            if p_value < alpha:
                print(f"\nSignificant effect found: {effect} (p = {p_value:.4f})")
            else:
                print(f"\nNo significant effect: {effect} (p = {p_value:.4f})")

        # **Visualization: Boxplot**
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=factor1, y=dependent_variable, hue=factor2, data=data)
        plt.title(f'Boxplot of {dependent_variable} by {factor1} and {factor2}')
        plt.xlabel(factor1)
        plt.ylabel(dependent_variable)
        plt.legend(title=factor2)
        plt.show()

        # **Visualization: Interaction Plot**
        from statsmodels.graphics.factorplots import interaction_plot
        plt.figure(figsize=(10, 6))
        interaction_plot(data[factor1], data[factor2], data[dependent_variable], colors=['red', 'blue', 'green'], markers=['D', 'o', '^'])
        plt.title(f'Interaction Plot of {dependent_variable} by {factor1} and {factor2}')
        plt.xlabel(factor1)
        plt.ylabel(dependent_variable)
        plt.show()

    except KeyError:
        print(f"Error: One or both factors '{factor1}' and '{factor2}' not found in the DataFrame.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example dataset
data = {'factor1': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'A', 'B', 'C'],
        'factor2': ['X', 'X', 'Y', 'X', 'X', 'Y', 'X', 'X', 'Y', 'Y', 'Y', 'Y'],
        'value':  [10, 12, 15, 18, 20, 22, 25, 27, 30, 14, 19, 28]}
df = pd.DataFrame(data)

# Run Two-Way ANOVA
perform_two_way_anova(df, 'value', 'factor1', 'factor2')
```

```python
#21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_f_distribution(df1_values, df2, x_max=5):

    x = np.linspace(0, x_max, 500)  # Generate x values

    plt.figure(figsize=(8, 5))

    # Plot F-distribution for different df1 values
    for df1 in df1_values:
        y = stats.f.pdf(x, df1, df2)  # Compute F-distribution PDF
        plt.plot(x, y, label=f"df1={df1}, df2={df2}")

    plt.title("F-Distribution for Different Degrees of Freedom")
    plt.xlabel("F-Value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

# Example: Visualize F-distribution for different df1 values with fixed df2=10
df1_values = [2, 5, 10]  # Different numerator degrees of freedom
df2 = 10  # Fixed denominator degrees of freedom

plot_f_distribution(df1_values, df2)
```

```python
#23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Simulate Data from a Normal Distribution
np.random.seed(42)  # Ensures reproducibility

# Generate two normal distributions (representing two groups)
group_A = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, Std Dev=10, n=100
group_B = np.random.normal(loc=55, scale=10, size=100)  # Mean=55, Std Dev=10, n=100

# Step 2: Perform Hypothesis Testing (Independent T-Test)
t_stat, p_value = stats.ttest_ind(group_A, group_B)

# Step 3: Interpret the Results
alpha = 0.05  # Significance level

print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The means are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in means.")

# Step 4: Visualize the Data Distribution
plt.figure(figsize=(8, 5))
sns.histplot(group_A, label="Group A", kde=True, color="blue", bins=20, alpha=0.6)
sns.histplot(group_B, label="Group B", kde=True, color="red", bins=20, alpha=0.6)
plt.axvline(np.mean(group_A), color="blue", linestyle="dashed", label=f"Mean A: {np.mean(group_A):.2f}")
plt.axvline(np.mean(group_B), color="red", linestyle="dashed", label=f"Mean B: {np.mean(group_B):.2f}")
plt.legend()
plt.title("Histogram of Simulated Data (Normal Distribution)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
import scipy.stats as stats


np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std Dev=10, n=30


sigma_0_squared = 100  # Hypothesized variance (e.g., std dev = 10, so variance = 10)


sample_variance = np.var(sample_data, ddof=1)  # Unbiased variance (ddof=1 for sample variance)
n = len(sample_data)  # Sample size

chi_square_stat = (n - 1) * sample_variance / sigma_0_squared

alpha = 0.05
p_value = 2 * min(
    stats.chi2.cdf(chi_square_stat, df=n-1),  # Lower tail probability
    1 - stats.chi2.cdf(chi_square_stat, df=n-1)  # Upper tail probability
)

# Step 6: Print Results
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 7: Interpretation
if p_value < alpha:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized variance.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in variance.")
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
import scipy.stats as stats

def z_test_proportions(successes_1, trials_1, successes_2, trials_2, alpha=0.05):
    """
    Perform a two-proportion Z-test.

    Parameters:
    - successes_1: Number of successes in group 1
    - trials_1: Total observations in group 1
    - successes_2: Number of successes in group 2
    - trials_2: Total observations in group 2
    - alpha: Significance level (default = 0.05)

    Returns:
    - Z-score and P-value
    """
    # Step 1: Calculate sample proportions
    p1 = successes_1 / trials_1
    p2 = successes_2 / trials_2

    # Step 2: Compute pooled proportion
    p_pool = (successes_1 + successes_2) / (trials_1 + trials_2)

    # Step 3: Compute standard error
    se = np.sqrt(p_pool * (1 - p_pool) * (1/trials_1 + 1/trials_2))

    # Step 4: Compute Z-score
    z_score = (p1 - p2) / se

    # Step 5: Compute P-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Print results
    print(f"Proportion 1 (p1): {p1:.4f}, Proportion 2 (p2): {p2:.4f}")
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference between the proportions.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between the proportions.")

# Example Usage
z_test_proportions(successes_1=120, trials_1=500, successes_2=100, trials_2=500)
```

```python
#26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def f_test_variance(data1, data2, alpha=0.05):
    """
    Perform an F-test for comparing the variances of two datasets.

    Parameters:
    - data1: First dataset (list or NumPy array)
    - data2: Second dataset (list or NumPy array)
    - alpha: Significance level (default = 0.05)

    Returns:
    - F-statistic and P-value with interpretation
    """

    # Step 1: Compute sample variances
    var1 = np.var(data1, ddof=1)  # Sample variance (unbiased)
    var2 = np.var(data2, ddof=1)  # Sample variance (unbiased)

    # Step 2: Compute F-statistic (larger variance on top)
    if var1 > var2:
        F = var1 / var2
        dfn, dfd = len(data1) - 1, len(data2) - 1  # Degrees of freedom
    else:
        F = var2 / var1
        dfn, dfd = len(data2) - 1, len(data1) - 1  # Degrees of freedom

    # Step 3: Compute P-value
    p_value = 2 * (1 - stats.f.cdf(F, dfn, dfd))  # Two-tailed test

    # Step 4: Print Results
    print(f"Variance of Dataset 1: {var1:.4f}")
    print(f"Variance of Dataset 2: {var2:.4f}")
    print(f"F-Statistic: {F:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Step 5: Interpretation
    if p_value < alpha:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in variances.")

    # Step 6: Visualization - Boxplot of Variances
    plt.figure(figsize=(8, 5))
    plt.boxplot([data1, data2], labels=['Dataset 1', 'Dataset 2'], patch_artist=True)
    plt.title("Boxplot of Variance Comparison")
    plt.ylabel("Values")
    plt.grid(True)
    plt.show()

# Example Usage: Simulate Random Data
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
data2 = np.random.normal(loc=55, scale=15, size=30)  # Mean=55, Std=15

f_test_variance(data1, data2)
```

```python
#27.  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
import scipy.stats as stats

def chi_square_goodness_of_fit(observed, expected, alpha=0.05):
    """
    Perform a Chi-Square Goodness of Fit test.

    Parameters:
    - observed: List of observed frequencies
    - expected: List of expected frequencies
    - alpha: Significance level (default = 0.05)

    Returns:
    - Chi-square statistic, p-value, and interpretation
    """

    # Step 1: Compute Chi-Square statistic and p-value
    chi_square_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

    # Step 2: Print Results
    print(f"Observed Frequencies: {observed}")
    print(f"Expected Frequencies: {expected}")
    print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Step 3: Interpretation
    if p_value < alpha:
        print("Reject the null hypothesis: The observed distribution is significantly different from the expected distribution.")
    else:
        print("Fail to reject the null hypothesis: The observed distribution follows the expected distribution.")

# Example Usage: Simulate Observed and Expected Data
np.random.seed(42)
observed_frequencies = np.random.randint(20, 50, size=5)  # Observed counts in 5 categories
expected_frequencies = np.full(5, np.mean(observed_frequencies))  # Expected uniform distribution

chi_square_goodness_of_fit(observed_frequencies, expected_frequencies)
```
