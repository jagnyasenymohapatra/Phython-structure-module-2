### **1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.**  

Data is categorized into **Qualitative** and **Quantitative**:  

- **Qualitative (Categorical) Data**: Non-numeric data used to categorize or label attributes.  
  - **Examples**: Gender (Male/Female), Eye color (Blue, Brown, Green), Marital Status (Single, Married)  
  - **Types**:  
    - **Nominal Scale**: No inherent order (e.g., Blood Group: A, B, AB, O).  
    - **Ordinal Scale**: Ordered but no meaningful difference between values (e.g., Education Level: High School, Bachelor’s, Master’s).  

- **Quantitative (Numerical) Data**: Represents numerical values with mathematical significance.  
  - **Examples**: Height (in cm), Weight (in kg), Salary (in dollars).  
  - **Types**:  
    - **Interval Scale**: Ordered, with equal intervals but no true zero (e.g., Temperature in Celsius).  
    - **Ratio Scale**: Ordered, with equal intervals and a true zero (e.g., Weight, Age, Income).

### **2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.**  

Measures of central tendency describe the "center" of a dataset:  

- **Mean (Arithmetic Average)**:  
  - Formula: \( \text{Mean} = \frac{\sum x_i}{n} \)  
  - **Use when**: Data is normally distributed (no extreme outliers).  
  - **Example**: Average income of employees in a company.  

- **Median (Middle Value)**:  
  - **Use when**: Data has outliers or skewed distribution.  
  - **Example**: If salaries in a company range from $10,000 to $1,000,000, the median gives a better central tendency than the mean.  

- **Mode (Most Frequent Value)**:  
  - **Use when**: Data is categorical or when identifying the most common occurrence is useful.  
  - **Example**: The most frequently sold shoe size in a store.  



### **3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?**  

**Dispersion** measures how spread out the data points are.  

- **Variance (σ²)**:  
  - Measures the average squared deviation from the mean.  
  - Formula: \( \sigma^2 = \frac{\sum (x_i - \bar{x})^2}{n} \)  
  - **Higher variance** means more spread out data.  

- **Standard Deviation (σ)**:  
  - Square root of variance, representing dispersion in the same unit as the data.  
  - Formula: \( \sigma = \sqrt{\sigma^2} \)  
  - **Example**: If test scores have a high standard deviation, students’ performances vary greatly.  



### **4. What is a box plot, and what can it tell you about the distribution of data?**  

A **Box Plot (Box-and-Whisker Plot)** is a graphical representation of data distribution using five key values:  
1. **Minimum (Q0)** – Smallest value (excluding outliers).  
2. **First Quartile (Q1)** – 25th percentile.  
3. **Median (Q2)** – 50th percentile.  
4. **Third Quartile (Q3)** – 75th percentile.  
5. **Maximum (Q4)** – Largest value (excluding outliers).  

It helps in:  
- Identifying **skewness** (if the median is closer to one end).  
- Detecting **outliers** (values outside whiskers).  
- Comparing **spread and variability**.  



### **5. Discuss the role of random sampling in making inferences about populations.**  

**Random Sampling** ensures that every individual in the population has an equal chance of being selected. It helps in:  
- **Reducing bias** and making results more generalizable.  
- **Improving accuracy** when estimating population characteristics.  
- **Enhancing reliability** in statistical analysis.  

**Example**: If we survey 100 randomly chosen students from a university, the results can be generalized to the entire student body.  


### **6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?**  

**Skewness** measures the asymmetry of data distribution:  

- **Positive Skew (Right-Skewed)**:  
  - Tail on the right, mean > median.  
  - **Example**: Income distribution (few people earn extremely high salaries).  

- **Negative Skew (Left-Skewed)**:  
  - Tail on the left, mean < median.  
  - **Example**: Age at retirement (most retire at 60-65, fewer retire early).  

- **Effect on Interpretation**:  
  - In **skewed data**, the mean is pulled towards the tail, making the median a better central tendency measure.  



### **7. What is the interquartile range (IQR), and how is it used to detect outliers?**  

**Interquartile Range (IQR)** measures data spread:  
- Formula: **IQR = Q3 - Q1**  
- **Use**: Identifies **outliers** using:  
  - **Lower Bound**: \( Q1 - 1.5 \times IQR \)  
  - **Upper Bound**: \( Q3 + 1.5 \times IQR \)  
- **Example**: In test scores, values outside this range are potential outliers.  


### **8. Discuss the conditions under which the binomial distribution is used.**  

A **Binomial Distribution** models the number of successes in \( n \) independent trials.  

**Conditions**:  
1. **Fixed number of trials (n)**  
2. **Only two outcomes (Success/Failure)**  
3. **Constant probability (p)**  
4. **Independent trials**  

**Example**: Tossing a coin 10 times (Success = Heads, Failure = Tails).  



### **9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).**  

**Normal Distribution** is a symmetric, bell-shaped curve defined by **mean (μ)** and **standard deviation (σ)**.  

**Empirical Rule (68-95-99.7 rule)**:  
- **68%** of data falls within **1σ** of the mean.  
- **95%** within **2σ**.  
- **99.7%** within **3σ**.  

**Example**: IQ scores follow a normal distribution with a mean of 100 and σ = 15. This means 95% of people have IQs between 70 and 130.  


### **10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.**  

A **Poisson Process** models rare events occurring in a fixed interval.  

**Example**: Suppose a call center receives an average of 5 calls per hour. The probability of receiving exactly 3 calls in an hour is:  

\[
P(X=k) = \frac{e^{-\lambda} \lambda^k}{k!}
\]

For \( k = 3 \), \( \lambda = 5 \):

\[
P(3) = \frac{e^{-5} 5^3}{3!} \approx 0.1404
\]

So, there's a **14.04% probability** of getting exactly 3 calls in an hour.  


### **11. Explain what a random variable is and differentiate between discrete and continuous random variables.**  

A **Random Variable (RV)** is a function that assigns numerical values to outcomes of a random process.  

- **Discrete RV**: Takes countable values (e.g., Number of students in a class).  
- **Continuous RV**: Takes an infinite range of values (e.g., Height of students).
